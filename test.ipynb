```python
from pathlib import Path

avgPath = Path('GFLPNCData/zh-CN/lua/AvgConfig')
locPath = Path('GFLPNCData/zh-CN/lua/LuaConfigs/locale_text.lua')
avgListPath = Path('GFLPNCData/zh-CN/lua/LuaConfigs/story_avg.lua')
charPath = Path('GFLPNCData/zh-CN/lua/LuaConfigs/avg_character.lua')
contentTypePath = Path('GFLPNCData/zh-CN/lua/Game/Avg/Enum/eAvgContentType.lua')

prop = {1:'Subtitle', 2:'name', 3:'name', 4:'name',5:'name'}

charTable = func.loadLuaFile(charPath)
charTable = charTable()

locTable = func.loadLuaFile(locPath)
locTable = locTable()

def reader(avgcfg, avglang):
    storyDict = {}
    idx = list(avgcfg)[0]

    while True:

        line = avgcfg[idx]
        lineDict = {}
        lineDict['id'] = f'node{idx}'
        lineDict['attributes'] = {}
        if line['contentType']:
            ct = line['contentType']
            lineDict['prop'] = prop[ct]
            if ct in (2,5):
                lineDict['attributes']['name'] = ''
            elif ct == 3:
                lineDict['attributes']['name'] = locTable[charTable[line['speakerHeroId']]['name']]
            elif ct == 4:
                lineDict['attributes']['name'] = avglang[line['speakerName']]
            
            lineDict['attributes']['content'] = avglang[line['content']]

        
        if line['branch']:
            lineDict['prop'] = 'Decision'
            values = [f"option{value}" for value in list(line['branch'])]
            options = ';'.join([avglang[line['branch'][value]['content']] for value in list(line['branch'])])
            lineDict['attributes']['values'] = values
            lineDict['attributes']['options'] = options
            lineDict['targetLine'] = {f"option{value}":line['branch'][value]['jumpAct'] for value in list(line['branch'])}

        storyDict[f'node{idx}'] = lineDict

        if line['nextId']:
            idx = line['nextId']
        elif line['isEnd']:
            break
        else:
            idx += 1

    return storyDict
```
